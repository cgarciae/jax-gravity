```
(* x = np.cos(theta) * np.sin(phi) *)
EqX := D[X[t] == R Cos[Theta[t]] Sin[Phi[t]], t]
EqX
```

```
(* y = np.sin(theta) * np.sin(phi) *)
EqY := D[Y[t] == R Sin[Theta[t]] Sin[Phi[t]], t]
EqY
```

```
(* z = np.cos(phi) *)
EqZ := D[Z[t] == R Cos[Phi[t]], t]
EqZ
```

```
Sol1 := Simplify[Solve[EqX && EqY, {Theta'[t], Phi'[t]}][[1]]]
SolTheta1 := Sol1[[1]]
SolPhi1 := Sol1[[2]]
```
